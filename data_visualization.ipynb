# Fetch Data

```python
from warnings import filterwarnings
filterwarnings("ignore")

import pandas as pd
A = pd.read_csv("C:/Users/Shruti/Downloads/Cars93.csv")
```

# Seperating categorical and continuous columns

```python
cat = []
con = []
for i in A.columns:
    if(A[i].dtypes == "object"):
        cat.append(i)
    else:
        con.append(i)
```

# Univariate plots(plot of 1 column)
    1. Continuous --> distribution plot, histogram(Spread of data)
    2. Categorical --> pie, HorBar char, VerBar chart

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
sb.distplot(A.Price)
```

```python
A.Price.hist()
```

```python
#bar:
A.Type.value_counts().plot(kind="bar")
plt.xlabel("Car Type")
plt.ylabel("No of cars")
plt.yticks(range(0,26,3))
```

```python
#horizontal_bar:
A.Type.value_counts().plot(kind="barh")
```

```python
#pile:
A.Type.value_counts().plot(kind="pie")
```

# Subplot

```python
plt.figure(figsize=(30,60))

x = 1

for i in A.columns:
    if(A[i].dtypes == "object"):
        plt.subplot(14,2,x)
        A[i].value_counts().plot(kind="barh")
        x = x + 1
    else:
        plt.subplot(14,2,x)
        sb.distplot(A[i])
        x = x+1
```
